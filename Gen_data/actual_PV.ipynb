*Filter the data with "Solar" production type.*

```python
import pandas as pd

data = pd.read_csv("gen_hourly_MW_2025.csv", parse_dates=['datetime'])#, index_col='datetime')
data.head()
```

```python
# Filter for Solar production type only
solar_data = data[data['Production Type'] == 'Solar'].reset_index(drop=True)

# Convert MW to MWh (1 hour Ã— MW = MWh)
solar_data['Generation (MWh)'] = solar_data['Generation (MW)']
solar_data = solar_data.drop('Generation (MW)', axis=1)

print(solar_data.head())
print(solar_data.tail())
```

```python
print("Solar Data Types:")
print(solar_data.dtypes)
print("Missing Values in Solar Data:")
solar_data.isna().sum()
```

```python
solar_data.to_csv("actual_PV_2025.csv", index=False)
```
